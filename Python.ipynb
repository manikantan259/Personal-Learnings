```python
class social:
  def syllabus(self):
    print("Hey Im social")

class math:
  def syllabus (self):
    print("Hey Im math")


def class_parcer(self):
  for i in self:
    i.syllabus()

socialObj= social()
mathObj = math()

listOf = [socialObj,mathObj]

class_parcer(listOf)
```

Encapsulation

```python
class atm :
  def  __init__ (self,balance):

    self.__balance = balance

  def deposit (self,amount):
    self.__balance = self.__balance+amount

  def withdraw(self,amount):
    if  self.__balance > amount :
      self.__balance =  self.__balance - amount
      return True
    else:
      return False



bank= atm(5000)

```

```python
bank.deposit(5000)
```

```python
bank.withdraw(20000)
```

```python
bank.withdraw(5000)
```

```python
class parent():
  def method1(self):
    print("this is parent")

class child(parent):
  def method2(self):
    print("this is child1")

class child1(child):
  def method3(self):
    print("this is child2")

class child2(child1,child,parent):
  pass
```

```python
child2_obj = child2()
```

Abstraction


```python
import abc

class pwskills :

    @abc.abstractmethod
    def student_details(self):
        pass

    @abc.abstractmethod
    def student_assignment(self):
        pass


    @abc.abstractmethod
    def student_marks(self):
        pass
```

```python
class data_science(pwskills):

    def student_details(self):
        return "it will try to return a details of data science masters "

    def student_assignment(self):
        return "it will return a details of student assignemnt for data science masters "
```

```python
ds = data_science()
ds.student_details()
```

Decorator

```python
def deco (fun):
  def inner():
    print("this is the start of fn")
    fun()
    print("this is end of fn")
  return inner
```

```python
@deco
def test():
  print(4+5)
```

```python
test()
```

```python
class learning:
    def __init__ (self,name, no,address):

      self.name = name
      self.no = no
      self.address = address
    @classmethod
    def method1(cls,name,no,address):
        return cls(name,no,address)
    def ret(self):
        return self.name ,self.no,self.address


ex = learning.method1("rohan",7447,"okokok")

ex.name
```

```python
class classMethod:
  mobile_no = 7337596569                                                     #class variable

  def __init__(self,name,color):                                               # method to pass variables
    self.name = name
    self.color=color


  @classmethod
  def change_mobile(cls,mobile):                                                  #classMethod
      classMethod.mobile_no=mobile

  @classmethod
  def details(cls,name,color):
    return cls (name,color)

  def ret(self):                                                                                  #instance Method
    print(self.name,self.color,classMethod.mobile_no)

```

```python
cls_obj = classMethod("mani","dark")
```

```python
cls_obj.ret()
```
