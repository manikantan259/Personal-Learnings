```python
#Two Sum


class Node:

    # Construct to create a new Node
    def __init__(self, key):

        self.data = key
        self.left = self.right = None

def insert(root: Node, key: int):

    # If the tree is empty, return a new Node
    if root is None:
        return Node(key)

    # Otherwise, recur down the tree
    if root.data > key:
        root.left = insert(root.left, key)

    elif root.data < key:
        root.right = insert(root.right, key)

    # return the (unchanged) Node pointer
    return root



def getInorderPath(root):
    # iterative approach
    stack = []
    currentNode = root
    path = []
    while True:
        if currentNode:
            stack.append(currentNode)
            currentNode = currentNode.left
        else:
            if not stack: # if len(stack) > 0:
                break
            currentNode = stack[-1]
            stack.pop()
            path.append(currentNode.data)
            currentNode = currentNode.right
    return path

    #-----------OR-------------

def inorder(root ,arr) :

    if (root == None):
        return

    inorder(root.left,arr)
    arr.append(root.data)

    inorder(root.right,arr)
    return arr



def twoSum(root,target):

  path = getInorderPath(root)
  left = 0
  right = len(path)-1

  while left<right:
    currentSum = path[left]+path[right]
    print(f"Pair Found: {path[left]} + {[right]} = {target}")

    if currentSum == target:
      return True

    elif currentSum > target:
      right -= 1

    else :
      left +=1

  print("No such values are found!")
  return False



if __name__ == "__main__":

    root = None
    root = insert(root, 15)
    root = insert(root, 10)
    root = insert(root, 20)
    root = insert(root, 8)
    root = insert(root, 12)
    root = insert(root, 16)
    root = insert(root, 25)

    twoSum(root, 33)
```

```python
# Python program to demonstrate delete operation
# in binary search tree

# A Binary Tree Node


class Node:

	# Constructor to create a new node
	def __init__(self, key):
		self.key = key
		self.left = None
		self.right = None


# A utility function to do inorder traversal of BST
def inorder(root):
	if root is not None:
		inorder(root.left)
		print (root.key,end=" ")
		inorder(root.right)


# A utility function to insert a
# new node with given key in BST
def insert(node, key):

	# If the tree is empty, return a new node
	if node is None:
		return Node(key)

	# Otherwise recur down the tree
	if key < node.key:
		node.left = insert(node.left, key)
	else:
		node.right = insert(node.right, key)

	# return the (unchanged) node pointer
	return node




def minValueNode(node):
	current = node

	# loop down to find the leftmost leaf
	while(current.left is not None):
		current = current.left

	return current

# Given a binary search tree and a key, this function
# delete the key and returns the new root


def deleteNode(root, key):


	if root is None:
		return root

	if key < root.key:
		root.left = deleteNode(root.left, key)

	elif(key > root.key):
		root.right = deleteNode(root.right, key)

	else:


		if root.left is None:

			return root.right

		elif root.right is None:

			return root.left


		temp = minValueNode(root.right)


		root.key = temp.key

		# Delete the inorder successor
		root.right = deleteNode(root.right, temp.key)

	return root


# Driver code
""" Let us create following BST
			50
		/	 \
		30	 70
		/ \ / \
	20 40 60 80 """

root = None
root = insert(root, 50)
root = insert(root, 30)
root = insert(root, 20)
root = insert(root, 40)
root = insert(root, 70)
root = insert(root, 60)
root = insert(root, 80)

print ("Inorder traversal of the given tree")
inorder(root)

print ("\nDelete 20")
root = deleteNode(root, 20)
print ("Inorder traversal of the modified tree")
inorder(root)

print ("\nDelete 30")
root = deleteNode(root, 30)
print ("Inorder traversal of the modified tree")
inorder(root)

print ("\nDelete 50")
root = deleteNode(root, 50)
print ("Inorder traversal of the modified tree")
inorder(root)


```

```python
#Construct A binary tree from preorder


class Node:
    def __init__(self, data):
        self.left = None
        self.right = None
        self.data = data


def constructBinarySearchTreeFromPreorder(preorder):
    idx = [0]
    upperBound = float("inf")
    return solve(preorder, idx, upperBound)

def solve(preorder, idx, ubValue):
    if idx == len(preorder) or preorder[idx[0]] > ubValue:
        return None
    newNode = Node(preorder[idx[0]])
    idx[0] += 1 # increasing the index by 1, index = index + 1
    newNode.left = solve(preorder, idx, newNode.data)
    newNode.right = solve(preorder, idx, ubValue)
    return newNode

```

```python
#Flatten BSt




def flattenBinary(root):


  if not root:
    return

  stack = [root]
  while stack : #while stack is not empty

    currentNode = stack.pop()

    if currentNode.right:
      stack.append(currentNode.right)

    if currentNode.left:
      stack.append(currentNode.left)
    if stack:
     currentNode.right = stack[-1]
     currentNode.left = None

  return root




```
